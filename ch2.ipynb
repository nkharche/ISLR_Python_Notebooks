## Chapter 2: Basic Operations
### Vectors, Data, Matrices and Subsetting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Array
# Initializing numpt arrays
x = np.array([2,7,5])
print (x)
y = np.arange(4,13,3)
print (y)
# Access elements by specifying offset from end
print (x[-2])
# Element indices can be specified as an array
print (x[np.array([0,1])])
```

```python
#Matrix
Z = np.matrix(np.arange(1,13))
print (Z)
Z = Z.reshape(4,3)
print (Z)
# note: Python (R) arranges the elements column(row)-wise
print (Z.shape)
# Flatten a matrix
print (Z.flatten())
```

### Generating Random Data, Graphics

```python
x = np.random.uniform(0., 1., 50)
print (x)
y = np.random.uniform(0., 1., 50)
print (y)
```

```python
fig, ax = plt.subplots()
plt.scatter(x, y, marker='o', color='r');
plt.xlabel("Random Uniform")
plt.ylabel("Random Normal")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2) # 1 row, 2 columns
ax1.scatter(x, y, marker='o', color='r');
ax2.hist(y);
```
